Question 1: Solutions to the Quadratic Equation

Write a program that computes the real solution(s) to the quadratic equation:

ax2+bx+c=0ax2+bx+c=0 
Your program should do the following:

Ask the user to enter the values of a, b and c (these should be of type float).
Check whether there are real solutions.
If there are real solution(s), compute and print them.
If there are not, inform the user.
Use at least one function from the module numpy.
Print the solution(s), if they exist.

```python
import numpy

def Quadratic_Equation():
    """ This function calculates solution to quadratic equation. """

    a = float(input("a ")) #input the value of a
    b = float(input("b ")) #input the value of b
    c = float(input("c ")) #input the value of c

    value = numpy.sqrt(b**2 - 4*a*c)

    if value >= 0:
      print("Solution to quadratic equation is:")

      first = -b + value 
      positive = first / 2*a

      second = -b - value 
      negative = second / 2*a  

      print('The positive is : ', positive, ' and the negative is :', negative) 

    else:
      print("Solution doesn\'t exist.")

Quadratic_Equation()      
```
