```
import requests
import pandas as pd
from bs4 import BeautifulSoup

htmlPage = requests.get('https://www.shopclues.com/seasonal-winter-wear.html')
soup = BeautifulSoup(htmlPage.content, 'html.parser')
product = soup.find(id='product_list')
print(len(product))
```

```
items = product.find_all(class_='column col4')
original_Price = product.find_all(class_='ori_price')

#productName, productPrice, productDiscount
name = [item.find(class_='prod_name').get_text() for item in items]
price = [price.find(class_='p_price').get_text() for price in original_Price]
discount = [price.find(class_='prd_discount').get_text() for price in original_Price]

print(name, price, discount)
```

```
records = pd.DataFrame({
  'Product Name': name,
  'Product Price': price,
  'Product Discount': discount,
})

records.to_csv('winterWearData.csv')
```
