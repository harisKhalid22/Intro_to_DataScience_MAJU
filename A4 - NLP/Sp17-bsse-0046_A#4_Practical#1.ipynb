# **Practical1: Write regular expressions (regex) to fulfil the following requirements:**

```
import re
```

```
#The simplest type of regex matches complete substrings in the same way as normal Python string processing.
```

```
#One word replacement using regex:

print('My name is haris.'.replace('haris', 'Haris Khalid'))
print(re.sub('haris', 'Haris Khalid', 'My name is haris.'), '\n')

#Advantages of regex with more than one replacement:

sample_data = 'My name is sheikh haris khalid.'
print(sample_data.replace('sheikh', 'Sheikh').replace('haris', 'Haris').replace('khalid', 'Khalid'))
print(re.sub('sheikh haris khalid', 'Sheikh Haris Khalid', sample_data))
```

```
text = 'Hello! My name is Sheikh Haris Khalid.'

print(re.findall('!', text), '\n')
print(re.findall('a', text), '\n')
print(re.sub('Hello', 'Assalamoallaikum', text), '\n')
re.split(' ', text)
```

```
print(re.findall('h\w', text), '\n')
re.findall('h\W', text)
```

```
sample_data = 'In computer programming, a comment is a programmer-readable explanation or annotation in the source code of a computer program. They are added with the purpose of making the source code easier for humans to understand, and are generally ignored by compilers and interpreters. The address is 1/100 - A Shah Faisal Colony, Karachi.'
```

```
#The sequence:

#prog = re.compile(pattern)
#result = prog.match(string)

#-------------------------

#which is equivalent to:

#result = re.match(pattern, string)
```

```
#Any character except for a new line

result = re.match('[\n]', sample_data)
print(result)
```

```
#A period

result = re.findall('[\.]', sample_data)
print(result)
```

```
#Any digit

result = re.findall('[\d]', sample_data)
print(result)
```

```
#Anything but a digit

result = re.findall('[^\d+]', sample_data)
print(result)
```

```
#Any character, including digits

result = re.findall('.', sample_data)
print(result)
```

```
#Anything but a character

result = re.findall('[^a-z, A-Z]', sample_data)
print(result)
```

```
#Collection of characters

result = re.findall('[a-z, A-Z]', sample_data)
print(result)
```

```
#Match something up to ‘n’ times

result = re.findall(r'\ba', sample_data)
print(result)
```

```
#Match 1 or more occurrences

result = re.findall(r'\ba+\w*', sample_data)
print(result)
```

```
#Match any number of occurrences (0 or more times)

result = re.findall(r'\ban*', sample_data)
print(result)
```

```
#Match exactly zero or one occurrence

result = re.findall(r'\ba*', sample_data)
print(result)
```

```
#Match word boundaries

result = re.findall('computer', sample_data)
print(result)
```

```
#Extract the user id, domain name and suffix from the following email addresses.

email = ''' awais@facebook.com, 
yasir@google.com, 
mazhar@amazon.com '''

result = re.findall(r'(\w*)@(\w*)\.(\w*)', email)
print(result)
```
