# FILE HANDLING

#                           <<<---create new file--->>>

```python
f = open("demofile.txt", "x")

#write on it
f = open("demofile.txt", "a")
f.write("Asalam e Alaikum !!")
f.close()
```

```python
try:
   f = open("demofile.txt",encoding = 'utf-8')
   for x in f:
      print(x)
finally:
   f.close()
```

##### Because file is now close no operation will going to perform on close file

```python
print(f.read())
```

#                            <<<-------Append data on file---->>>

```python
f = open("demofile.txt", "a")
f.write("Now the file has more content!")
f.close()

#open and read the file after the appending:
f = open("demofile.txt", "r")
print(f.read())
```

#                                <<<----Overwrite file data----->>>>

```python
f = open("demofile.txt", "w")
f.write("Woops! I have deleted the content!")
f.close()

#open and read the file after the appending:
f = open("demofile.txt", "r")
print(f.read())
```

#                                <<<---Deleting the file--->>>

```python
import os
if os.path.exists("demofile.txt"):
  os.remove("demofile.txt")
else:
  print("The file does not exist")
```

# Class and Objects

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("Fahad", 22)

print(p1.name)
print(p1.age)
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def myfunc(self):
    print("Hello my name is " + self.name)

p1 = Person("Fahad", 22)
p1.myfunc()
```

# Inheritence 

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def printname(self):
    print(self.name, self.age)
    
x = Person("fahad", 22)
x.printname()

class Person2(Person):
  pass

x = Person2("Anas", 20)
x.printname()
```

# Operator overloading

```python
class OpOverload: 
    def __init__(self, x, y): 
        self.x = x 
        self.y = y 
  
    def __add__(self, other): 
        return self.x + other.x, self.y + other.y 
  
    def __str__(self): 
        return self.x, self.y 
  
Ob1 = OpOverload(1, 2) 
Ob2 = OpOverload(2, 3) 
Ob3 = Ob1 + Ob2 
print(Ob3) 
```
